#Generando datos aleatorios de manera simples con numpy

```python
import numpy as np
import pandas as pd
np.random.seed(423)
```

```python
a_buscar = ['clasica', 'rock', 'jazz', 'funk', 'pop', 'kpop', 'techno', 'drumnbass']
```

```python
acceso = np.random.choice(a_buscar,30000)
```

```python
todos_accesos = pd.DataFrame(acceso, columns= ["URI"])
```

```python
todos_accesos.URI = '/playlist/' + todos_accesos.URI
```

```python
paginas = ['/promocion_primer_de_abril/compra/plan_basico','/compra/plan_basico','/compra/plan_avanzado',
           '/','/musicas','/contacto']
```

```python
prob = np.array([1,1,1,10,4,2])
```

```python
accesos = np.random.choice(paginas,150000, p = prob/sum(prob))
```

```python
accesos = pd.DataFrame(accesos,columns=["URI"])
```

```python
todos_accesos = pd.concat([accesos,todos_accesos])
```

```python
def cero_nadie(id):
  if id<=100:
    return None
  return id-100
```

```python
todos_accesos['Usuario registrado'] = ((np.random.rand(len(todos_accesos))) * (len(todos_accesos) / 10)).round().astype(int)
```

```python
todos_accesos['Usuario registrado'] = todos_accesos['Usuario registrado'].apply(cero_nadie)
```

```python
status = [200,500]
```

```python
todos_accesos['Status de respuesta'] = np.random.choice(status,p = [0.97, 0.03] )
```

```python
todos_accesos = todos_accesos.sample(frac=1)
```

```python
todos_accesos.head()
```

```python
todos_accesos['URI'].value_counts()
```

```python
todos_accesos.to_csv("todos_accesos.csv",index=False)
```
